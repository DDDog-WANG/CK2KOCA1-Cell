# 染色種類

## クラス1 ~ クラス4

* 488は転写共役因子`CTCF`を染色したものです。CTCFは`３次元的にゲノム領域を分けるboundaryに集積する`（TADの境界に集積する）タンパク質です。

* 555はヒストンH3タンパク質の９番目のリジン残基がトリメチル化されたもの`（H3K9me3）`を染色しております。H3K9me3は`ヘテロクロマチン`に集積します。

## クラス5 ~ クラス6

* 488はエンハンサー領域に集積する`H3K4me1`となっております。
* 555は核膜に存在する`LaminB1`となっております。

```python
# define gamma function
def create_gamma_img(gamma, img):
    gamma_cvt = np.zeros((256,1), dtype=np.uint8)
    for i in range(256):
        gamma_cvt[i][0] = 255*(float(i)/255)**(1.0/gamma)
    return cv2.LUT(img, gamma_cvt)
```

```python
# see the origenal image
fig, ax = plt.subplots(2,2, figsize = (20,10))
n = 0
while n < 36:
    img = DataList_imread[n]
    plt.subplot(6,6,n+1)
    plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
    n = n + 1
plt.show()
```

```python
# see the gamma changes
fig, ax = plt.subplots(2,2, figsize = (20,10))
n = 0
while n < 36:
    img = DataList_imread[n]
    img_gamma = create_gamma_img(10, img)
    plt.subplot(6,6,n+1)
    plt.imshow(cv2.cvtColor(img_gamma, cv2.COLOR_BGR2RGB))
    n = n + 1
plt.show()
```

```python
# adjust all image with different gamma 
from tqdm import tqdm

for num in tqdm(range(36)):
    img = DataList_imread[num]
    for gamma in range(1,11):
        img_gamma = create_gamma_img(gamma, img)

        savename = "./Datagamma/" + DataLabelList[num] + "_gamma" + str(gamma) + ".tif"
        # print("save as: ", savename)
        cv2.imwrite(savename, img_gamma)
```

```python
for num in tqdm(range(24)):
    img = DataList_imread[num]
    gamma = 10
    img_gamma = create_gamma_img(gamma, img)
    savename = "./Databest/" + DataLabelList[num] + "_gamma" + str(gamma) + ".tif"
    cv2.imwrite(savename, img_gamma)
    
for num in tqdm(range(24,36)):
    img = DataList_imread[num]
    gamma = 6
    img_gamma = create_gamma_img(gamma, img)
    savename = "./Databest/" + DataLabelList[num] + "_gamma" + str(gamma) + ".tif"
    cv2.imwrite(savename, img_gamma)
```
